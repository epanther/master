```python
import pandas as pd
import numpy as np
df = pd.read_csv('/workspaces/master/train.csv')
```

```python
numeric_stats = df.describe()
print(numeric_stats)
```

```python
# Example: Average purchase amount by gender
average_purchase_by_gender = df.groupby('Gender')['Purchase'].mean()
print(average_purchase_by_gender)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Purchase', bins=30, kde=True)
plt.title('Histogram of Purchase Amounts')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Purchase')
plt.title('Box Plot of Purchase Amounts by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Product_Category_1', hue='Gender')
plt.title('Count of Product Categories by Gender')
plt.xlabel('Product Category 1')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Product_Category_1', hue='Gender')
plt.title('Count of Product Categories by Gender')
plt.xlabel('Product Category 1')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Age', palette='Set1')
plt.title('Distribution of Customer Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Your DataFrame and other code here

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='City_Category', hue='Gender')
plt.title('Count of Customers by City Category and Gender')
plt.xlabel('City Category')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
