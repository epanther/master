```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of the visualization
sns.set(style="whitegrid")

# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Purchase', bins=30, kde=True)
plt.title('Histogram of Purchase Amounts')
plt.xlabel('Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Purchase')
plt.title('Box Plot of Purchase Amounts by Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Product_Category_1', hue='Gender')
plt.title('Count of Product Categories by Gender')
plt.xlabel('Product Category 1')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()

```
